## Vizzy Sample Notebook (Vizzy_Token)

The data needed for this tutorial can be downloaded at https://www.kaggle.com/datasets/namanj27/ner-dataset

Our first step is to get our data ready to be turned into a Vizzy object. Vizzy objects accept data as a Pandas DataFrame, so we will need to read in our CSV's as DataFrames

```python
import pandas as pd
with open('data/ner_datasetreference.csv', encoding="utf8", errors='ignore') as f:
    df = pd.read_csv(f)[0:100]
```

Let's take a look at our data

```python
df.head()
```

Import the vizzy package library

```python
from vizzy import vizzy
```

The vizzy package has three types of objects: vizzy_doc, vizzy_sentence, and vizzy_token

Vizzy_tokens takes in two required inputs: a Pandas DataFrame and column as a string value. Vizzy_tokens is meant to also include a third input- 'label', which is the name of the column which contains the labels of your text. This 'label' input is optional. The labels argument must be a list of columns.

For our demo, we will look at the 'Word' column and the 'POS' column

```python
viz = vizzy.vizzy_token(dataframe = df, text = 'Word', split = 'dep', labels = ['shape', 'Tag', 'POS', 'lemma'])
```

Vizzy Sentence objects only have a few functions to help you vizualise your data. We will go through the in the cells below

```python
'''show_labels_count() shows a histogram of your data. The input for this function is the column you want
to see'''
viz.show_labels_count('POS')
```

```python
'''The show_labels_count_split() function will show you a bar graph of your tokens 
split by the "split" input you gave when initializing the object. This arguments for this function are the values
you want to split by given as a list'''
viz.show_labels_count_split(splits=['prep', 'ROOT', 'nsubj'])
```

```python
'''viz.show_case() shows your tokens broken down into upper, lower, and integer in a bar graph'''
viz.show_case()
```

```python
'''viz.print_case shows you the same data, but printed'''
viz.print_case()
```

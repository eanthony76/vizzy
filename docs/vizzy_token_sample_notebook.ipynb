The vizzy package has three types of objects: vizzy_doc, vizzy_sentence, and vizzy_token

Vizzy_tokens takes in two required inputs: a Pandas DataFrame and column as a string value. Vizzy_tokens is meant to also include a third input- 'label', which is the name of the column which contains the labels of your text. This 'label' input is optional, but currently there are no useable functions for vizzy_tokens objects without this input. You can only include one 'label' column per vizzy_token object.

For our demo, we will look at the 'Word' column and the 'POS' column

```python
viz = vizzy.vizzy_token(df, 'Word', 'POS')
```

Vizzy Sentence objects only have a few functions to help you vizualise your data. We will go through the in the cells below
